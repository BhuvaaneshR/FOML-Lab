```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df=pd.read_excel("/logistic_regression.xlsx")
df
```

```python
plt.scatter(df.CGPA,df.Placement)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[["CGPA"]],df.Placement,test_size=0.2)
```

```python
x_test
```

```python
y_test
```

```python
x_train
```

```python
LR=linear_model.LogisticRegression()
LR.fit(x_train,y_train)
```

```python
LR.predict(x_test)
```
